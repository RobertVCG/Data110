```python
#Task 1

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


offenses = [
    "Aggravated Assault",
    "Burglary",
    "Larceny",
    "Motor Vehicle Theft",
    "Murder",
    "Rape (revised)",
    "Robbery"
]

rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]



plt.figure(figsize=(10, 6))
sns.barplot(x=rates_2023, y=offenses)


plt.title('Crime Rates per 100,000 People (2023)', fontsize=16)
plt.xlabel('Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Crime', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()

```

```python
#Task 2

offenses = [
    "Aggravated Assault",
    "Burglary",
    "Larceny",
    "Motor Vehicle Theft",
    "Murder",
    "Rape (revised)",
    "Robbery"
]


changes = [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]

plt.figure(figsize=(10, 6))
sns.barplot(x=changes, y=offenses, dodge=False)


plt.title('Change in Crime Rates (2022 vs. 2023)', fontsize=16)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Crime', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

This bar graph makes more sense than Task 1's because percentage changes show how crime rates are increasing or decreasing over time, while absolute numbers can be misleading since they dont reveal whether crimes are getting better or worse.

```python
# version 1 add data labels

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=changes, y=offenses)


for index, value in enumerate(changes):
    bar_plot.text(value, index, f"{value:+.1f}", color='black', ha='left', va='center', fontsize=10)


plt.title('Change in Crime Rates (2022 vs. 2023)', fontsize=16)
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14)
plt.ylabel('Type of Crime', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Version 2 use different font size and style.

plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x=changes, y=offenses)


for index, value in enumerate(changes):
    bar_plot.text(value, index, f"{value:+.1f}", color='black', ha='left', va='center', fontsize=12, fontweight='bold')

plt.title('Change in Crime Rates (2022 vs. 2023)', fontsize=18, fontweight='bold')
plt.xlabel('Change in Offenses per 100,000 People', fontsize=14, fontstyle='italic')
plt.ylabel('Type of Crime', fontsize=14, fontstyle='italic')
plt.grid(axis='x', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

```python
#Optional Task


offenses = [
    "Aggravated Assault",
    "Burglary",
    "Larceny",
    "Motor Vehicle Theft",
    "Murder",
    "Rape (revised)",
    "Robbery"
]

rates_2022 = [273.1, 272.9, 1416.5, 284.4, 6.5, 42.1, 67.0]
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]


plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(offenses))


plt.barh(index, rates_2022, bar_width, label='2022', color='b')


plt.barh(index + bar_width, rates_2023, bar_width, label='2023', color='g')


plt.xlabel('Offenses per 100,000 People')
plt.ylabel('Type of Crime')
plt.title('Crime Rates Comparison: 2022 vs. 2023')
plt.yticks(index + bar_width / 2, offenses)
plt.legend()


plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
