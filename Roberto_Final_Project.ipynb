```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'https://raw.githubusercontent.com/RobertVCG/Data110/refs/heads/main/Dataset/adult_tobacco_consumption_2000_present_cdc.csv'
df = pd.read_csv(file_path)
```

```python
yearly_trends = df.groupby('Year')[['Domestic', 'Imports', 'Total']].sum()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(yearly_trends.index, yearly_trends['Total'], color='blue', marker='o', linewidth=2)

# Chart styling
plt.title("Total Tobacco Consumption Over Time (2000-2023)")
plt.xlabel("Year")
plt.ylabel("Total Consumption (Units)")
plt.grid(True)
plt.show()
plt.figure(figsize=(10, 6))
```

This graph shows the trend of total tobacco consumption (Domestic + Imports) from 2000 to 2023, highlighting whether consumption is increasing or decreasing.



```python
last_5_years = df[df['Year'] >= df['Year'].max() - 4]
domestic_vs_imports = last_5_years.groupby('Year')[['Domestic', 'Imports']].sum()

domestic_vs_imports.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'orange'])

plt.title("Domestic vs. Imported Tobacco Consumption (Last 5 Years)")
plt.xlabel("Year")
plt.ylabel("Consumption (Units)")
plt.legend(["Domestic", "Imports"])
plt.xticks(rotation=0)
plt.show()
```

This bar chart compares Domestic and Imported tobacco consumption for the last 5 years, making it easy to see the differences.



```python
latest_year = df['Year'].max()
latest_data = df[df['Year'] == latest_year].groupby('Year')[['Domestic', 'Imports']].sum()

plt.figure(figsize=(8, 8))
plt.pie(latest_data.iloc[0], labels=['Domestic', 'Imports'], autopct='%1.1f%%', startangle=140, colors=['skyblue', 'orange'])

plt.title(f"Domestic vs. Imported Tobacco Consumption in {latest_year}")
plt.show()
```

This pie chart shows the proportion of Domestic and Imported consumption for the most recent year, clearly visualizing the breakdown.



```python
avg_per_capita = df.groupby('Year')['Total Per Capita'].mean()


plt.figure(figsize=(12, 6))
avg_per_capita.plot(kind='bar', color='purple')

plt.title("Average Tobacco Consumption Per Person by Year")
plt.xlabel("Year")
plt.ylabel("Per Capita Consumption (Units)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```

This bar chart highlights the average tobacco consumption per person over the years, making it easy to track changes in usage per capita.



```python

yearly_consumption = df.groupby('Year')[['Domestic', 'Imports']].sum()


plt.figure(figsize=(12, 6))
plt.plot(yearly_consumption.index, yearly_consumption['Domestic'], label='Domestic Consumption', color='skyblue', linewidth=2, marker='o')
plt.plot(yearly_consumption.index, yearly_consumption['Imports'], label='Imported Consumption', color='orange', linewidth=2, marker='o')


plt.title("Domestic vs. Imported Tobacco Consumption (2000-2023)")
plt.xlabel("Year")
plt.ylabel("Consumption (Units)")
plt.legend()
plt.grid(True)
plt.show()
```

This graph compares Domestic and Imported consumption as two lines over the years, showing how each has changed relative to the other.
