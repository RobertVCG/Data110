```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/RobertVCG/Data110/refs/heads/main/Dataset/Daily_Arrests_20241001%20(1).csv')
data.head()
```

```python
theft_df = data[data['OFFENSE'].str.contains("theft", case=False, na=False)]
theft_df.head()
```

```python
total_theft_incidents = theft_df.shape[0]
age_range = (theft_df['AGE'].min(), theft_df['AGE'].max())
print(f"Total Theft Incidents: {total_theft_incidents}")
print(f"Age Range: {age_range}")
```

Total Theft Incidents: 96

Age Range: (18, 67)

```python
plt.figure(figsize=(10, 6))
theft_df['AGE'].hist(bins=15, edgecolor='black')
plt.title('Age Distribution of Theft Offenders')
plt.xlabel('Age')
plt.ylabel('Number of Offenders')
plt.tight_layout()
plt.show()


age_group = theft_df['AGE'].value_counts().idxmax()
age_group_count = theft_df['AGE'].value_counts().max()
print(f"The most common age group involved in theft offenses is {age_group} years old with {age_group_count} offenders.")
```

The most common age group involved in theft offenses is 26 years old with 11 offenders.

```python
offense_counts = theft_df['OFFENSE'].value_counts()

plt.figure(figsize=(12, 8))
offense_counts.plot(kind='barh')
plt.title('Most Common Types of Theft Offenses')
plt.xlabel('Number of Incidents')
plt.ylabel('Type of Theft Offense')
plt.tight_layout()
plt.show()

print("Top Theft Offense Types:")
print(offense_counts.head())
```
