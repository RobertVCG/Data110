```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


titanic = sns.load_dataset('titanic')


titanic.head()
```

The Titanic dataset includes details about 891 passengers on the ship. It has information like:

survived: Whether they lived (1) or not (0).

pclass: Their ticket class (1st, 2nd, or 3rd).

sex: Whether they were male or female.

age: Their age.

sibsp: Number of siblings or spouses with them.

parch: Number of parents or children with them.

fare: How much they paid for the ticket.

embarked: Where they boarded the ship.

Other columns show if they were alone, their deck, and if they survived. This data helps analyze factors that influenced survival.

```python
plt.figure(figsize=(10, 6))


sns.histplot(titanic['fare'], kde=True, bins=30)
plt.title('Distribution of Fares Paid by Titanic Passengers')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.grid(True)


plt.show()
```


Most people paid less than 50 for their tickets, which were mostly third-class. A few paid over 500, likely first class passengers. People who paid more had a better chance of surviving because they probably had better access to lifeboats.

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset('titanic')


plt.figure(figsize=(10, 6))

sns.kdeplot(data=titanic, x='fare', hue='survived', common_norm=False)
plt.title('Fare Distribution for Survivors and Non-Survivors')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(True)


plt.show()
```

People who paid more for their tickets were more likely to survive. The plot shows that survivors mostly paid higher fares, while non-survivors paid lower fares. This suggests that wealthier passengers who could afford higher fares had better chances of surviving, likely because they had more access to lifeboats.

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset('titanic')

plt.figure(figsize=(12, 6))

sns.histplot(data=titanic, x='fare', hue='pclass', kde=True, bins=30)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.grid(True)

plt.show()

plt.figure(figsize=(12, 6))
sns.kdeplot(data=titanic, x='fare', hue='pclass', multiple="stack")
plt.title('Survival by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.grid(True)

plt.show()
```

Fare Patterns: First-class passengers paid much higher fares than second and third class passengers. Most third-class passengers paid much less, reflecting how ticket prices were based on wealth.

Survival Patterns: First-class passengers had better survival rates, likely because they were wealthier and had more access to lifeboats. Third class passengers, who paid less, had lower survival chances, showing the clear division between rich and poor on the Titanic.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(10, 6))

sns.kdeplot(data=titanic, x='age', hue='survived', common_norm=False)
plt.title('Age Distribution of Survivors and Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)

plt.show()
```

Trends: Survivors tended to be younger, with many children surviving. Non-survivors had a wider range of ages, including more older passengers.

Impact: Younger people, especially children, had a better chance of surviving. This might be because children and women were given priority during the evacuation.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

male_survivors = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 1)]
female_survivors = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 1)]
male_non_survivors = titanic[(titanic['sex'] == 'male') & (titanic['survived'] == 0)]
female_non_survivors = titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 0)]

plt.figure(figsize=(12, 6))

plt.hist(male_survivors['age'], bins=20, alpha=0.5, label='Male Survivors', orientation='horizontal', color='blue')
plt.hist(male_non_survivors['age'], bins=20, alpha=0.5, label='Male Non-Survivors', orientation='horizontal', color='lightblue')

plt.hist(female_survivors['age'], bins=20, alpha=0.5, label='Female Survivors', orientation='horizontal', color='red')
plt.hist(female_non_survivors['age'], bins=20, alpha=0.5, label='Female Non-Survivors', orientation='horizontal', color='pink')

plt.title('Pyramid Histogram of Age, Survival, and Sex')
plt.xlabel('Number of Passengers')
plt.ylabel('Age')
plt.legend(loc='upper right')
plt.grid(True)

plt.show()
```

Trends: More women, especially younger ones, survived compared to men. Many older men didnt survive.

Impact: Women and children were given priority in lifeboats, so more females and younger people had a better chance of surviving.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='fare', data=titanic)
plt.title('Fare by Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Age vs. Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

Bar Graph: First class passengers had a much higher survival rate than those in second or third class.

Line Plot: There's no clear pattern between age and fare, but younger passengers generally paid lower fares.

Scatter Plot: Survivors tend to cluster around higher fares and middle age, showing that wealth and age could both influence survival.
