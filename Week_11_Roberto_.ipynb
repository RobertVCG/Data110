```python
import pandas as pd
import folium
from folium.plugins import HeatMap

bpd_arrests_data = pd.read_csv('BPD_Arrests.csv')

df = pd.read_csv('BPD_Arrests.csv')

df.head()

print("Columns:", df.columns)

print("Missing values:\n", df.isnull().sum())

df_clean = df.dropna(subset=['Latitude', 'Longitude', 'Gender'])

male_data = df_clean[df_clean['Gender'].str.lower() == 'male']
female_data = df_clean[df_clean['Gender'].str.lower() == 'female']

base_map = folium.Map(location=[df_clean['Latitude'].mean(), df_clean['Longitude'].mean()], zoom_start=12)

male_heat_data = list(zip(male_data['Latitude'], male_data['Longitude']))
HeatMap(male_heat_data, radius=10, name='Male Arrests').add_to(base_map)

female_heat_data = list(zip(female_data['Latitude'], female_data['Longitude']))
HeatMap(female_heat_data, radius=10, gradient={0.4: 'blue', 0.65: 'lime', 1: 'red'}, name='Female Arrests').add_to(base_map)

folium.LayerControl().add_to(base_map)

base_map.save('gender_based_heatmap.html')
base_map
```
