```python
import numpy as np
import matplotlib.pyplot as plt
print(plt.style.available)


def f(x):
    return x**2 - 2*x + 1

x = np.linspace(-2, 4, 400)


y = f(x)

plt.plot(x, y, label='f(x) = x^2 - 2x + 1')


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 - 2x + 1 over [-2, 4]')
plt.grid(True)
plt.legend()


plt.show()
```

```python
def f(x):
    return x**3

def g(x):
    return np.exp(x)

def h(x):
    return np.log(x)

x_values = np.linspace(0.1, 3, 500)

y_f = f(x_values)
y_g = g(x_values)
y_h = h(x_values)


plt.figure(figsize=(8, 6))
plt.plot(x_values, y_f, label='f(x) = x^3', color='blue')
plt.plot(x_values, y_g, label='g(x) = e^x', color='green')
plt.plot(x_values, y_h, label='h(x) = log(x)', color='red')

plt.legend()

plt.title('Plots of f(x) = x^3, g(x) = e^x, and h(x) = log(x)')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```

```python
def f(x):
    return x**2

def g(x):
    return x**3

def h(x):
    return x**4

def i(x):
    return x**5

x_values = np.linspace(-2, 2, 500)


y_f = f(x_values)
y_g = g(x_values)
y_h = h(x_values)
y_i = i(x_values)


plt.figure(figsize=(8, 6))


plt.plot(x_values, y_f, label='f(x) = x^2', color='blue', linestyle='-')
plt.plot(x_values, y_g, label='g(x) = x^3', color='#FF5733', linestyle='--')
plt.plot(x_values, y_h, label='h(x) = x^4', color=(0.1, 0.6, 0.1), linestyle='-.')
plt.plot(x_values, y_i, label='i(x) = x^5', color='C2', linestyle=':')


plt.legend()


plt.title('Plots of f(x) = x^2, g(x) = x^3, h(x) = x^4, and i(x) = x^5')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True)
plt.show()

```

```python
def f(x):
    return 1 / (x**2 + 1)


x_values = np.linspace(-10, 10, 500)


y_f = f(x_values)


plt.figure(figsize=(8, 6))
plt.plot(x_values, y_f, label='f(x) = 1 / (x^2 + 1)', color='blue')
plt.title('Plot of f(x) = 1 / (x^2 + 1) with Custom Axes Limits')
plt.xlabel('x')
plt.ylabel('y')

plt.xlim(-10, 10)
plt.ylim(0, 1.2)

plt.legend()
plt.grid(True)
plt.show()


```

```python
def f(x):
    return np.sqrt(x)

def g(x):
    return x**(1/3)

x_values = np.linspace(0, 10, 500)


y_f = f(x_values)
y_g = g(x_values)


plt.figure(figsize=(8, 6))


plt.plot(x_values, y_f, label='f(x) = sqrt(x)', color='blue')
plt.plot(x_values, y_g, label='g(x) = x^(1/3)', color='green')

plt.legend()

plt.title('Plots of f(x) = sqrt(x) and g(x) = x^(1/3)')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y, 'o', markersize=8)
plt.title('Scatter Plot using plt.plot()')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(x, y, color='blue', marker='o')
plt.title('Scatter Plot using plt.scatter()')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 4 * np.pi, 100)

y = x * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='orange', linewidth=2, label='f(x) = x sin(x)')
plt.scatter(x, y, marker='o', s=100, facecolor='blue', edgecolor='red', linewidth=1.5)

plt.title('Scatter Plot of f(x) = x sin(x)', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)

y = x * np.cos(x)

plt.figure(figsize=(10, 6))

plt.plot(x, y, color='green', linestyle='--', linewidth=2, label='f(x) = x cos(x)')

scatter_x = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
scatter_y = scatter_x * np.cos(scatter_x)
plt.scatter(scatter_x, scatter_y, color='red', marker='o', s=100, edgecolor='black', linewidth=1.5, label='Scatter Points')

plt.title('Plot of f(x) = x cos(x) with Scatter Points', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
def f(x):
    return np.tan(x)

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = f(x)


y_limit = 10


plt.figure(figsize=(12, 6))


plt.plot(x, y, color='orange', linestyle='-', linewidth=2, label='f(x) = tan(x)')


scatter_x = np.arange(-2 * np.pi, 2 * np.pi + np.pi/4, np.pi/4)
scatter_y = f(scatter_x)


valid_scatter_x = scatter_x[np.abs(scatter_y) < y_limit]
valid_scatter_y = scatter_y[np.abs(scatter_y) < y_limit]
plt.scatter(valid_scatter_x, valid_scatter_y, color='blue', marker='o', s=100, edgecolor='black', label='Scatter Points', zorder=5)

plt.title('Custom Plot of f(x) = tan(x)', fontsize=18)
plt.xlim(-2 * np.pi, 2 * np.pi)
plt.ylim(-y_limit, y_limit)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True)
plt.xlabel('X-axis (radians)', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.legend()
plt.tight_layout()


plt.show()
```

In this plot it helps you see the tangent functions includin the ups and down, and the points where it goes to infinity, the scatter points help highlight specific values to make the graph easier to understand.


```python
def i(x):
    return np.arctan(x)

x = np.linspace(-10, 10, 1000)
y = i(x)
plt.figure(figsize=(12, 6))

plt.plot(x, y, color='purple', linestyle='-', linewidth=2, label='i(x) = arctan(x)')

scatter_x = np.arange(-10, 11, 1)
scatter_y = i(scatter_x)

plt.scatter(scatter_x, scatter_y, color='orange', marker='o', s=100, edgecolor='black', label='Scatter Points')

plt.xlim(-10, 10)
plt.ylim(-np.pi/2, np.pi/2)

plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')

plt.grid(True)

plt.title('Plot of i(x) = arctan(x)', fontsize=18)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

plt.legend()

plt.tight_layout()
plt.show()
```

This plot shows how the arctangent function behaves as x increases and decreases, the scatter points highlight specific values making the graph easier to understand.
